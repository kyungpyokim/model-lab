### 함수로 만들기

```python
def chatbot5(user_msg, system_msg): # noqa
    res = client.responses.create( # noqa
    model="gpt-5-nano",
    input=[ # type: ignore
        {'role': 'system', 'content': system_msg },
        {'role': 'user', 'content': user_msg }
    ])

    return res.output_text
```

```python
system_prompt = '''
    당신은 AI 관련 개념을 아주 쉽게 설명해주는 인공지능 전문가입니다.
    사용자가 질문하는 개념에 대해 300자 이내로 비유를 통해 설명해주세요.
'''

user_msg = 'Reasoning Model이 뭔가요?'

res_txt = chatbot5(user_msg, system_prompt)
print(res_txt)
```

### 대화를 기억하는 챗봇 만들기

```python
def mybot(user_msg, conversation_id):
    system_prompt = '개그를 잘하는 개발자. 아재개그를 섞어서 대답하기'

    res = client.responses.create( # noqa
        model="gpt-5-nano",
        input=[ # noqa
            {'role': 'user', 'content': user_msg },
            {'role': 'system', 'content': system_prompt }
        ],
        conversation=conversation_id
    )

    return res.output_text

```

```python
from mybot import mybot

while True:
    msg = input('메시지 입력')
    if msg == '그만':
        break

    answer = mybot(msg)

    print(f'[나]: {user_msg}')
    print(f'[지피티]: {answer}')
```
