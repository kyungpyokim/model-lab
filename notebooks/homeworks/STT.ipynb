```python
import whisper

from lib.utils.path import audio_path

model = whisper.load_model('base')
path = str(audio_path() / 'meeting.mp3')

result = model.transcribe(path, language='ko', fp16=True)

print(result)
print(f'인식된 텍스트: {result["text"]}')
```

```python
import json

def convert_stt_to_markdown(data):
    # JSON 문자열인 경우 파싱, 딕셔너리인 경우 그대로 사용
    if isinstance(data, str):
        data = json.loads(data)

    segments = data.get('segments', [])

    md_output = []
    md_output.append("# 📝 회의 녹취록 정리\n")
    md_output.append("## 🕒 타임라인 및 대화 내용\n")

    for segment in segments:
        start_time = segment['start']
        # 시간을 MM:SS 형식으로 변환
        minutes = int(start_time // 60)
        seconds = int(start_time % 60)
        timestamp = f"{minutes:02d}:{seconds:02d}"

        text = segment['text'].strip()

        # 마크다운 리스트 형식으로 추가
        md_output.append(f"* **[{timestamp}]** {text}")

    # md_output.append("\n---\n")
    # md_output.append("## 📌 주요 안건 및 결정 사항\n")
    # md_output.append("* **디자인 팀:** 목요일까지 1차 수정본 제출 예정")
    # md_output.append("* **개발 팀:** 기능 개발 80% 완료, 다음 주 월요일부터 내부 테스트 시작")
    # md_output.append("* **마케팅 팀:** 기획 초안 제출 기한 변경 (다음 주 수요일)")
    # md_output.append("* **차기 회의 일정:** 4월 29일(월) 오전 11시")

    return "\n".join(md_output)

# 입력 데이터 (제공해주신 데이터)
stt_data = {
    'text': result['text'],
    'segments': [
        {'id': 0, 'start': 0.0, 'text': ' 다들 도착하셨나요? 그럼 회의 시작할게요.'},
        # ... (이하 생략)
    ]
}



# 실행 및 결과 확인
# 실제 사용 시에는 변수에 전체 데이터를 넣으시면 됩니다.
markdown_result = convert_stt_to_markdown(result)
print(markdown_result)
```
